## Activity 4.04: Visualizing the Top 30 Music YouTube Channels Using Seaborn's FacetGrid
In this activity, we will visualize the total number of subscribers and the total number of views for the top 30 YouTube channels (as of January 2020) in the category music by using the FacetGrid() function that's provided by the Seaborn library. Visualize the given data using a FacetGrid with two columns. The first column should show the number of subscribers for each YouTube channel, whereas the second column should show the number of views. The goal of this activity is to get some practice working with FacetGrids. Following are the steps to implement this activity:

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Use pandas to read the youtube.csv data located in the Dataset folder.

```python
mydata = pd.read_csv("../../Datasets/youtube.csv")
```

```python
mydata.head()
```

Access the data of each group in the column, convert them into a list, and assign this list to variables of each respective group.

```python
channels = mydata[mydata.columns[0]].tolist()
```

```python
print(channels)
```

```python
subs = mydata[mydata.columns[1]].tolist()
```

```python
views = mydata[mydata.columns[2]].tolist()
```

```python
print(subs)
```

```python
print(views)
```

Create a pandas DataFrame with the preceding data, using the data of each respective group.

```python
data = pd.DataFrame({'YouTube Channels': channels + channels, 'Subscribers/Views in millions': subs + views, 
                     'Type': ['Subscribers'] * len(subs) + ['Views'] * len(views)})
```

Visualize the given data using a FacetGrid with two columns. The first column should show the number of subscribers for each YouTube channel, whereas the second column should show the number of views.

```python
sns.set()
g = sns.FacetGrid(data, col='Type', hue='Type', sharex=False, height=8)
g.map(sns.barplot, 'Subscribers/Views in millions', 'YouTube Channels')
plt.show()
```
